### 1.2 Álgebra Lineal Aplicada
----
A partir del dataset, crear arreglos de precios y cantidades. Calcular el vector de ingresos brutos (multiplicación matricial).

```python
import numpy as np

techventas_dataset = "../data/ventas_techventas.csv"

# Obtener las columnas de cantidad y precio unitario
columnas_a_utilizar = (7, 8)

codificacion_del_dataset = "latin-1"

# Cargar el dataset de techventas
columna_cantidad, columna_precio = np.genfromtxt(
    techventas_dataset, 
    delimiter=',', 
    encoding=codificacion_del_dataset, 
    skip_header=1, 
    usecols=columnas_a_utilizar,
    dtype=None,
    unpack=True
)

print("Arreglo de cantidades (Con valor absoluto aplicado): ")
print(columna_cantidad)
print()
print("Arreglo de precios unitarios: ")
print(columna_precio)
```

```python
# Obtener el vector de ingresos brutos  por medio de la multiplicación matricial de cantidad y precio

ingresos_brutos = np.dot(columna_cantidad, columna_precio)
print(f"Ingresos brutos de Techventas: {ingresos_brutos}")
```

#### **Aclaración del ejercicio**
Teniendo en cuenta que la columna de cantidad contiene valores negativos, estos modifican los ingresos brutos totales y las estadísticas descriptivas del punto 1.3 del ejercicio, y a pesar que se podría aplicar valor absoluto para convertir estos valores en positivos con np.abs, ese procedimiento le correspondería a la sección de limpieza con Pandas, donde se pueden hacer las transformaciones o la eliminación de los registros con datos incorrectos y en general definir si se puede modificar ese dato o se filtra en su lugar.

### 1.3 Estadísticas con Numpy
----
Usar Numpy para calcular la media, mediana, desviación estándar, y percentiles 25 y 75 de la columna de precios. Posteriormente analizar la información que brindan estos datos respecto al catálogo.

```python
# Estadísticas descriptivas
media = np.mean(columna_precio)
mediana = np.median(columna_precio)
desviacion_estandar = np.std(columna_precio)
percentiles = np.percentile(columna_precio, [25, 75])

print("Medidas de tendencia central:")
print(f" Media: {media} \n Mediana: {mediana} \n")
print("Medidas de dispersión:")
print(f" Desviación estándar: {desviacion_estandar} \n percentil 25%: {percentiles[0]} \n percentil 75%: {percentiles[1]}")
```

#### Análisis de los datos obtenidos:
1. La media: Permite obtener el valor típico dentro del grupo de datos. En este caso indica el precio que se puede esperar de un producto dado (378.25), considerando que puede ser alterado por los extremos.

2. La mediana: Es el valor que se encuentra en el medio del grupo de datos, e indica un valor típico más realista (120) que la media, dado que no se ve afectado por los extremos.

3. Desviación estándar: Indica cuánto se aleja en promedio, un precio individual respecto a la media. Una baja desviación indica una alta consistencia en los precios, mientras que una alta desviación, una amplia variedad. En este caso, siendo la desviación estándar tan alta, implica que en el dataset existen unos pocos productos con precios considerablemente más elevados que el resto.

4. Percentil 25 y 75: Permiten dividir el grupo de datos en cuartiles. El 50% de las ventas de Techventas ocurren entre el perccentil 25 y 75, el resto de los datos corresponden a artículos más económicos o más costosos y que no corresponden a las ventas típicas que se pueden esperar dentro de la empresa.
